```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first', sparse_output=False)

from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

from sklearn.pipeline import Pipeline
```

```python
from sklearn.compose import  ColumnTransformer
preprocessor = ColumnTransformer([
    ('ohe', ohe, ['Placement_Training', 'Gender', 'Placement_Status']),
    ('scale', scale, ['CGPA','Internship','Paid_Internship','Projects','Workshop_Certification','Aptitude_Score','Skills_Rating','SSC_Marks'])
    
], remainder='passthrough')

main_preprocessing_pipeline = Pipeline(['preprocessor', preprocessor])
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=5)
```

```python
x = df.drop(columns=['Placement_Status'])
y = df['Placement_Status']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=42)

from sklearn.metrics import accuracy_score
```

```python
x_train_trf = main_preprocessing_pipeline.fit_transform(x_train, y_train)
x_test_trf = main_preprocessing_pipeline.transform(x_test)
```
